```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import randint
from numpy import mean
```

```
results = []
for num_trials in range(1,1000):
    trials = np.random.poisson(11, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})

df.plot()
```

```
means = [mean(np.random.poisson(11, 1000)) for _ in range(1000)]


plt.hist(means)
plt.show()
```

```
results = []
for num_trials in range(1,1000):
    trials = np.random.normal(10, 20, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})


df.plot()
```

```
means = [mean(np.random.normal(10, 20, 1000)) for _ in range(1000)]

plt.hist(means)
plt.show()
```
